```python
%matplotlib inline
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
# Load in csv
pyber_ride_df = pd.read_csv("Resources/PyBer_ride_data.csv")
pyber_ride_df
```

```python
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)")
plt.show()
```

```python
# Set x-axis and tick locations.
x_axis = np.arange(len(pyber_ride_df))
tick_locations = [value for value in x_axis]
# Plot the data.
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)")
plt.xticks(tick_locations, pyber_ride_df["Month"])
plt.show()
```

```python
pyber_ride_df.plot.bar(x="Month", y="Avg. Fare ($USD)")
plt.show()
```

```python
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)", kind='bar', color="skyblue")

plt.show()
```

```python
# Import dependencies.
import matplotlib.pyplot as plt
import statistics
pyber_ride_df
```

```python
stdev=statistics.stdev(pyber_ride_df["Avg. Fare ($USD)"])
```

```python
stdev
```

```python
pyber_ride_df.plot.bar(x="Month", y="Avg. Fare ($USD)", color="skyblue")
plt.errorbar(pyber_ride_df["Month"], pyber_ride_df["Avg. Fare ($USD)"], yerr=stdev, capsize=3, errorevery=2, elinewidth=None, capthick=2, barsabove=True)
plt.yticks(np.arange(0, 55, step=5.0))
fig, ax = plt.subplots()
plt.show()

```
